```python
import numpy as np 
```

```python
# 驗證公式計算是否為20分貝
20*np.log10(200/20)
```

```python
#1.正常的談話的聲壓為20000微巴斯卡，請問多少分貝?
#請寫下程式
V1 = 20000
V0 = 20
20*np.log10(V1/V0)
```

```python
#2.30分貝的聲壓會是50分貝的幾倍?
#公式移項過後可以得到 V1 = ?
#請寫下程式
```

```python
# 20*np.log10(V1/20) = 30
# np.log(V1/20) = 30/20
# V1/20 = 10**(30/20)
# V1 = (10**(30/20)) * 20
V1_30 = (10**(30/20)) * 20
V1_50 = (10**(50/20)) * 20
ans = round(V1_30 / V1_50 , 2)
ans
```
